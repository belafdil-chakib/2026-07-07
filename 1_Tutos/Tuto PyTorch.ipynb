Comme NumPy, il fournit des tenseurs multidimensionnels et une large gamme d’opérations rapides,
mais il ajoute aussi le support GPU et l’autograd pour l’optimisation.

Copie

```python
a = torch.tensor([1,2,3])
b = a  # Copie superficielle
b[0] = 100
print(a)  # a est modifié
```

```python
# Si on veut réaliser une copie, utiliser clone 
a = torch.tensor([1,2,3])
b = a.clone()  # Copie
b[0] = 100
print(a)  # a n'est pas modifié
print(b)
```

Statistiques

```python
stats = torch.tensor([[1,2,3],[4,5,6]])
print(stats)
print(torch.min(stats))
print(torch.max(stats, dim=1).values)
print(torch.sum(stats, dim=0))
```

Réorganiser les tableaux

```python
before = torch.tensor([[1,2,3,4],[5,6,7,8]])
print(before)
after = before.reshape((8,1))
print(after)
```

```python
v1 = torch.tensor([1,2,3,4])
v2 = torch.tensor([5,6,7,8])
print(torch.vstack([v1,v2,v1,v2]))
```

```python
h1 = torch.ones((2,4))
h2 = torch.zeros((2,2))
print(torch.hstack([h1,h2]))
```

Charger des données depuis un fichier

```python
import numpy as np

filedata = torch.tensor(
    np.genfromtxt(
        'https://raw.githubusercontent.com/KeithGalli/NumPy/master/data.txt',
        delimiter=',',
        dtype='int32'
        )
)
    
print(filedata)
print(filedata.dtype)
```

Masquage booléen

```python
filedata.shape
```

```python
print(filedata[~((filedata > 50) & (filedata < 100))])
```

Indexer avec une liste

```python
a = torch.arange(1,10)
print(a)
print(a[torch.tensor([1,2,8])])
```

Conditions

```python
print(torch.any(filedata > 50))
print(torch.all(filedata >= 0))
```
