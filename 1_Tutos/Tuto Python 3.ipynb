### Fonctions avec plusieurs sorties

```python
def mean_std(x1, x2, x3, x4, x5):
    mean = (x1 + x2 + x3 + x4 + x5) / 5
    std = (((x1 - mean)**2 + (x2 - mean)**2 + (x3 - mean)**2 + (x4- mean)**2 + (x5 - mean)**2) / 5) ** 0.5
    return mean, std
```

```python
m1, std1 = mean_std(1, 1, 1, 1, 1)
print(f"moyenne = {m1} | std1 = {std1}")

m1, std1 = mean_std(0, 0, 1, 2, 2)
print(f"moyenne = {m1} | std1 = {std1}")
```

### On peut réécrire la fonction qui peut maintenant s'adapter à la taille de la liste

```python
def mean_std(numbers):
    # Calcul de la moyenne
    total = 0
    for number in numbers:
        total += number
    mean = total / len(numbers)
    
    # Calcul de l'écart-type
    total = 0
    for number in numbers:
        total += (number - mean) **2
    std = (total / len(numbers)) ** 0.5
    
    return mean, std
```

```python
m1, std1 = mean_std([1, 1, 1, 1, 1])
print(f"moyenne = {m1} | std1 = {std1}")

m2, std2 = mean_std([0, 0, 1, 2, 2])
print(f"moyenne = {m2} | std1 = {std2}")

m3, std3 = mean_std([0, 1, 2])
print(f"moyenne = {m3} | std1 = {std3}")
```

# Les boucles while et instruction break

### la cellule avec une boucle for ci-dessous

```python
total = 0
for num in range(5):
    total += num # incrémentation
print(total)
```

### peut être réécrite avec une boucle while

```python
total = 0
j = 1
while j < 5:
    total += j
    j += 1
print(total)
```

```python
# Imaginons le but est de calculer la somme des éléments positifs d'une liste ordonnée (décroissante)
list_numbers = [5, 4, 4, 3, 1, -2, -3, -5]

i = 0
sum_num = 0
while list_numbers[i] > 0:
    sum_num += list_numbers[i]
    i += 1
print(sum_num)
```

```python
# Peut être écrit avec une boucle for et une instruction break
sum_num = 0
for num in list_numbers:
    if num < 0:
        break
    sum_num += num
print(sum_num)
```

```python
# Ou avec une boucle while et une instruction break
sum_num = 0
i = 0
while True:
    if list_numbers[i] < 0:
        break
    sum_num += list_numbers[i]
    i += 1
print(sum_num)
```

# Classes et objets

## Création d'une classe Robot ayant pour attribut un nom, une couleur et un poids. Une méthode de classe permettra au robot de se présenter.

```python
class Robot:
    def __init__(self, name, color, weight):
        self.name = name
        self.color = color
        self.weight = weight
    def introduce_self(self):
        print(f"Mon nom est {self.name}")
```

```python
r1 = Robot("C-3PO", "dorée", 75)
r2 = Robot("R2-D2", "blanc", 32)
```

```python
print(r1.name)
print(r2.weight)
```

```python
r1.introduce_self()
r2.introduce_self()
```
