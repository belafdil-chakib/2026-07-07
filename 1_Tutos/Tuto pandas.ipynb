<I>pandas</I> fournit des structures de données et des fonctions de haut niveau conçues pour rendre le travail avec des données structurées ou tabulaires rapide, facile et expressif. Depuis son émergence en 2010, il a contribué à faire de Python un environnement d'analyse de données puissant et productif.
Les principaux objets de pandas qui seront utilisés ici sont le <I>DataFrame</I>, une structure de données tabulaire orientée colonne avec des étiquettes de ligne et de colonne, et la <I>Series</I> un objet tableau étiqueté unidimensionnel.

<I>pandas</I> associe les idées de calcul de tableau hautes performances de <I>NumPy</I> aux capacités flexibles de manipulation de données des feuilles de calcul et des bases de données relationnelles (telles que SQL). Il fournit des fonctionnalités d'indexation sophistiquées pour faciliter le remodelage, le découpage, la réalisation d'agrégations et la sélection de sous-ensembles de données.

## Agréger / transformer (GroupBy)

```python
df.groupby(['Type 1', 'Type 2']).agg({'Name': 'count'})
```

```python
df['count'] = df.groupby(['Type 1', 'Type 2'])['Name'].transform('count')
df.head(100)
```

```python
df.groupby('Type 1').agg({'Defense': 'mean'}).sort_values(by='Defense', ascending=False)
```

```python
def delta_min_max(x):
    return x.max() - x.min()

df.groupby('Type 1').agg({'Defense': delta_min_max}).sort_values(by='Defense', ascending=False)
```

```python
df.groupby('Type 1').boxplot(column=['Defense', 'Attack'], figsize=(21, 12))
```

## Travailler avec de grandes quantités de données



```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.txt', sep='\t', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
```
